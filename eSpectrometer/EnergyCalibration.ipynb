```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

(Energy, PositionA00) = np.loadtxt("A00.txt", unpack=True)
(_Energy, PositionA04) = np.loadtxt("A04.txt", unpack=True)
(_Energy, PositionA08) = np.loadtxt("A08.txt", unpack=True)
(_Energy, PositionB04) = np.loadtxt("B04.txt", unpack=True)
(_Energy, PositionB08) = np.loadtxt("B08.txt", unpack=True)
(_Energy, PositionB12) = np.loadtxt("B12.txt", unpack=True)
```
