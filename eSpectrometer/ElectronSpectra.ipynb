```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import mahotas as mh
from wavelets import WaveletAnalysis

class Shot(object):
    
    def __init__(self, Date, Run, Number):
        self.Date = Date
        self.Run = Run
        self.Number = Number

def buildDirectoryName(Path, Date, Item):
    DirectoryName = Path + "\\" + Date + "\\" + Item + "\\"
    if Item in ["ProbeCamX", "Delay", "CamX", "Y", "Z"]:
        DirectoryName = Path + "\\" + Date + "\\" + "Probe" + "\\"        
    return DirectoryName

def buildFileName(Path, Shot, Item):
    Directory = buildDirectoryName(Path, Shot.Date, Item)
    FileName = Directory + Item + ".txt"
    if Item in ["ePointing", "eSpecHighE", "eSpecLowE", "eSpecSlow",  "Probe"]:
        FileName = Directory + Shot.Date + "_" + Shot.Run + "_" + Shot.Number + "_" + Item + ".png"
    return FileName
```

```python
# 20161201_027_6363_probe.png
path = "O:\\Electrons\\"
date = "20161201"
run = "008"
shotNo = "2306"
item = "ePointing"

shot = Shot(date, run, shotNo)
image = mh.imread(buildFileName(path, shot, item))



height = 50
width = 50

top = 360
left =210
imageROI = np.double(image[top:top+height,left:left+width])
imageBKG = np.double(image[1:1+height,1:1+width])

figROI = plt.figure()
plt.imshow(imageROI)
plt.show()

figBKG = plt.figure()
plt.imshow(imageROI-imageBKG)
plt.show()
print(np.mean(image))
print(np.mean(imageBKG))
print(np.sum(imageBKG)/np.sum(imageROI))
```
